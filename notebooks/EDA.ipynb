### SVO distribution

```python
svo = df[df.svo.notnull()]
```

```python
# Style settings
sns.set_style('whitegrid')

def remove_gridlines(ax):
    ax.yaxis.grid(False)
    ax.xaxis.grid(False)
    
def modify_gridlines(ax):
    for line in ax.get_ygridlines():
        line.set_linestyle(':')

    for line in ax.get_xgridlines():
        line.set_linestyle(':')    

def add_svo_boundaries(ax, text=False):
    plt.axvline(-12.04, color='gray', linewidth=.5)
    plt.axvline(22.45, color='gray', linewidth=.5)
    plt.axvline(57.15, color='gray', linewidth=.5)

    if text:
        ax.text(-22.04, .125, 'Individualistic', fontsize=10, color='gray')
        ax.text(16, .125, 'Prosocial', fontsize=10, color='gray')    
        ax.text(50, .125, 'Altruistic', fontsize=10, color='gray')
```

```python
# SVO distribution
ax = plt.axes()
modify_gridlines(ax)
add_svo_boundaries(ax, True)
sns.distplot(svo.svo, bins=30)
# plt.savefig('../img/svo_dist.png')
```

The overall distribution looks multimodal, with the primary mode around 35 degress (in the prosocial category) and the other around 6 or 7 degress (in the individualistic) case

```python
ax = plt.axes()
modify_gridlines(ax)
add_svo_boundaries(ax)

sns.distplot(svo[svo.type=='prosocial'].svo, bins=30)
sns.distplot(svo[svo.type=='individualistic'].svo, bins=30)
#sns.distplot(svo[svo.type=='competitive'].svo, bins=30)
plt
```

The distribution looks to be multi-modal, with a primary and secondary mode for each of the main SVO types.  The primary points could be from either default or attraction points in the interface (places where it's easy to move sliders) or they could reflect real attraction points for the different types

### Clustering

```python
cluster_cols = [
    'age', 'income',
    'education', 'politics',
    'religiosity', 'trust_development',
    'relationship_development',
    'dependence_development',
]
```

```python
interest_cols = [
    'health_fitness', 'indoor',
    'life_style', 'nature', 'outdoor',
    'performance_arts', 'sports', 'other'
]
```

```python
event_cols = [
    'ev0', 'ev1', 'ev2',
    'ev3', 'ev4', 'ev5',
    'i0', 'i1', 'i2', 'i4'
]
```

```python
svo = df.loc[
    (df.svo.notnull()) & 
    (df.income!= 'Prefer not to answer') & 
    (df.ip != '104.200.153.94') &
    (df.indoor.notnull())
, :]
svo.loc[:, 'income'] = svo.loc[:, 'income'].astype(int)
svo.loc[:, event_cols] = svo.loc[:, event_cols].fillna(0)
```

```python
svo.loc[:, interest_cols] = svo[interest_cols] / 5
```

```python
target_cols = cluster_cols
```

```python
cluster_df = svo.loc[:, ['type'] + target_cols]
```

```python
from sklearn.manifold import TSNE
```

```python
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
```

```python
X = cluster_df.ix[:, 1:].as_matrix()
X_scaled = min_max_scaler.fit_transform(X)
```

```python
Y = cluster_df.type.as_matrix()
```

### TSNE

```python
model = TSNE(n_components=2, perplexity=50)
```

```python
fit_values = model.fit_transform(X_scaled)
```

```python
clustered_df = pd.DataFrame(fit_values)
```

```python
df = pd.concat([clustered_df, cluster_df.type], axis=1)
```

```python
# tdf.columns = ['x', 'y', 'z', 'type']
df.columns = ['x', 'y', 'type']
```

```python
plt.figure(figsize=(12, 12))
sns.lmplot(x='x', y='y', hue='type', fit_reg=False, data=df, scatter_kws={'alpha':0.5})
```
