### Helper Functions


```python
def remove_gridlines(ax):
    ax.yaxis.grid(False)
    ax.xaxis.grid(False)
    
def modify_gridlines(ax):
    for line in ax.get_ygridlines():
        line.set_linestyle(':')

    for line in ax.get_xgridlines():
        line.set_linestyle(':')

def add_svo_boundaries(ax, text=False):
    plt.axvline(-12.04, color='gray', linewidth=.5)
    plt.axvline(22.45, color='gray', linewidth=.5)
    plt.axvline(57.15, color='gray', linewidth=.5)

    if text:
        ax.text(-36, .125, 'Competitive', fontsize=10)
        ax.text(-4, .125, 'Individualistic', fontsize=10)
        ax.text(32, .125, 'Prosocial', fontsize=10)    
        ax.text(63, .125, 'Altruistic', fontsize=10)
```

```python
def angle2cat(SVO_angles):
    if SVO_angles >= 57.15:
        return 2 #"altruistic"
    elif SVO_angles < 57.15 and SVO_angles >= 22.45:
        return 0 #"prosocial"
    elif SVO_angles < 22.45 and SVO_angles >= -12.04:
        return 1 #"individualistic"
    elif SVO_angles < -12.04:
        return 3 #"competitive"
```

```python
def plot_tsne(X, n_components, perplexity):
    model = TSNE(n_components=n_components, perplexity=perplexity)
    fit_values = model.fit_transform(X)
    clustered_df = pd.DataFrame(fit_values)
    tdf = pd.concat([clustered_df, cluster_df.type], axis=1)
    tdf.columns = ['x', 'y', 'type']
    plt.figure(figsize=(12, 12))
    sns.lmplot(x='x', y='y', 
               hue='type', fit_reg=False, 
               data=tdf, scatter_kws={'alpha':0.7})
```
